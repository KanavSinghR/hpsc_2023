__My Exponential__ <br />
In the below cell, exponential function will be created.

```python
'''
This function computes the exponential of a number using the exponential series
'''
import numpy as np
def myexp(x,terms=100,convg=False,tol=10**(-10)):
    '''
    myexp(x,terms=n)
    x=Number whose exponential is to be calculated
    terms=Number of terms to be used in exponential series
    convg=Shows the term at which our function converge to np.exp for some tolerance
    '''
    summation=1.0;
    assert terms>=0,'Number of terms must be positive'
    fact=1.0;
    for i in range (1,terms):
        fact=fact*i
        summation=summation+x**i/fact
        if abs(summation-np.exp(x))<tol:#Convergence is checked within the function to avoid additional calculations.
            if convg==True:
                print(f"Our function converged at term {i}")
            break
        elif i==terms-1:
            if convg==True:
                print(f"Our function did not converge for {i+1} terms")
    return summation
```

```python
myexp(2)
```

__Function Plot__ <br />
In the below cell, we will plot our exponential function simultaneously with numpy.exp function. 

```python
import matplotlib.pyplot as plt
for x in np.linspace(0,100,1000):
    '''
    #linspace is taken in for loop because when "if abs(summation-np.exp(x)<tol)" is taken, summation is a single 
    valued variable whereas, np.exp become a multivalued variable(in our case array) and therefore cannot be 
    compared.
    '''
    plt.plot(x,myexp(x), color='red', marker='.')
    plt.plot(x,np.exp(x), color='blue', marker='*')
plt.legend(["exp","myexp"])
plt.xlabel("XValues")
plt.ylabel("YValues")
plt.title("myexp vs. exp plot")
plt.rcParams["figure.figsize"] = (10,6)#To increase size of plot
plt.show()
```
