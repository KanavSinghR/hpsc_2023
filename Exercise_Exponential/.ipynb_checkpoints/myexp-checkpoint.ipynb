__My Exponential__ <br />
In the below cell, exponential function will be created.

```python
'''
This function computes the exponential of a number using the exponential series
'''
def myexp(x,terms=100,convg=False,tol=10**-14):
    '''
    myexp(x,terms=n)
    x=Number whose exponential is to be calculated
    terms=Number of terms to be used in exponential series
    '''
    summation=1.0;
    assert terms>=0,'Number of terms must be positive'
    fact=1.0;
    for i in range (1,terms):
        fact=fact*i
        summation=summation+x**i/fact
    if convg==True:
        from numpy import exp
        if abs(myexp(x,terms=terms)-exp(x))<tol:
            print("Our function converges (for tolerance value: %20.15f)"%tol)
        else:
            print("Our function does not converge (for tolerance value: %20.15f)"%tol)
    return(summation)
```

__Function Plot__ <br />
In the below cell, we will plot our exponential function simultaneously with numpy.exp function. 

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(0,100,1000)
plt.plot(x,myexp(x))
plt.plot(x,np.exp(x))
plt.legend(["exp","myexp"])
plt.xlabel("XValues")
plt.ylabel("YValues")
plt.title("myexp vs. exp plot")
```

__Time Comparison__ <br />
In the below cell, we will compare the time taken by myexp and exp function for some random value.

```python
import random
y=random.uniform(-100,100)
%timeit myexp(y)
%timeit np.exp(y)
```
